```julia
Setup = "B[256]"

t = 1.0
U = 20.0
V = 3.0
Lx = 256
δ = 0.3
β = 100.0

AF = GetPointFS(
    "AF",
    t,U,V,
    Lx,δ,β,
    "./FSFinder-AF-$(Setup).txt",
    "/home/nepero27178/Thesis/LadderHubbardSC/Project/HartreeFock/processing/Phase=AF/fermi-surface/"
)
FakeAF = GetPointFS(
    "FakeAF",
    t,U,V,
    Lx,δ,β,
    "./FSFinder-FakeAF-$(Setup).txt",
    "/home/nepero27178/Thesis/LadderHubbardSC/Project/HartreeFock/processing/Phase=FakeAF/fermi-surface/"
)
"done"
```

```julia
xx = AF.Curve[1]["kx"]./pi
yy = AF.Curve[1]["ky"]./pi
scatter(xx,yy,label="AF",markerstrokewidth=0,aspect_ratio=:equal,size=(700,700))

xx = FakeAF.Curve[1]["kx"]./pi
yy = FakeAF.Curve[1]["ky"]./pi
scatter!(xx,yy,markerstrokewidth=0,label="FakeAF")
title!("U=$(U), V=$(V), δ=$(δ)")
xlabel!("kx/pi")
ylabel!("ky/pi")
savefig("./FS_U=$(U)_V=$(V)_δ=$(δ).png")
```
