```julia
using DataFrames
using Plots
using DelimitedFiles
using ColorSchemes

FilePathIn = "/home/nepero27178/Thesis/LadderHubbardSC/Project/HM-HartreeFock/simulations/special/fixed-Ut-scan/vary-U/AF.txt"
Data = readdlm(FilePathIn, ';')
DF = DataFrame(Data[2:end,:], Data[1,:])
"done"
```

```julia
xx = unique(DF.x)
 = unique(DF.)

P = plot(xlabel="x", ylabel="m", legend=:outertopright)
for (j,) in enumerate()
    df = DF[DF..==,:]
    yy = [eval.(Meta.parse.(df.v[j]))["m"] for j in 1:length(df.x)]
    plot!(df.x, yy, color=ColorSchemes.lipari25[2*j], label="=$()")
end
title!("U=10, t=U/x")
P
```

```julia
P = plot(xlabel="x", ylabel="/t", legend=:outertopright)
for (j,) in enumerate()
    df = DF[DF..==,:]
    yy = df. ./ df.t
    plot!(df.x, yy, color=ColorSchemes.lipari25[2*j], label="=$()")
end
title!("U=10, t=U/x")
P
```

```julia
P = plot(xlabel="x", ylabel="(-)/t", legend=:outertopright)
for (j,) in enumerate()
    df = DF[DF..==,:]
     = [eval.(Meta.parse.(df.v[j]))["m"] for j in 1:length(df.x)] .* 10
    yy = df. ./ df.t
    plot!(df.x, yy, color=ColorSchemes.lipari25[2*j], label="=$()")
end
title!("U=10, t=U/x")
P
```
