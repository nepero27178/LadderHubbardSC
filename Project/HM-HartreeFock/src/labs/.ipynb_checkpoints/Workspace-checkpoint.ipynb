```julia
using Plots
using DelimitedFiles
using DataFrames
using ColorSchemes

ROOT_FOLDER = @__DIR__
include(ROOT_FOLDER * "/../../../EHM-HartreeFock/src/setup/graphic-setup.jl")
include(ROOT_FOLDER * "/../../../EHM-HartreeFock/src/modules/methods-simulating.jl")
```

```julia
Parameters::Dict{String,Float64} = Dict([
    "t" => 1.0,
    "U" => 10.0,
    "V" => 0.0
])

Lx = 128
L = [Lx, Lx]
δ = 0.15
β = 100.0

p = 100
Δv::Dict{String,Float64} = Dict([
    "m" => 1e-4,
    "w0" => 1e-4,
    "wp" => 1e-4
])
Δn = 1e-2
g = 0.5

Results = RunHFAlgorithm(
    "AF",Parameters,L,0.5+δ,β,
    p,Δv,Δn,g;
    debug=true,
    verbose=true
)
Results
```

```julia
v = Results[1]
μ = Results[4]
print(v["m"]*10-μ)
Kx::Vector{Float64} = [kx for kx in -1:2/Lx:1]
popfirst!(Kx)
Ky::Vector{Float64} = [ky for ky in -1:2/Lx:1]
popfirst!(Ky)
K::Matrix{Vector{Float64}} = [ [kx,ky] for kx in Kx, ky in Ky ]

Nk = GetKPopulation("AF",Parameters,K,v,μ,β)
```
