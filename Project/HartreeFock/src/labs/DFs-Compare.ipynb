```julia
using Plots
using DelimitedFiles
using DataFrames

RootFolder = "/home/nepero27178/Documents/University/Masters/LadderHubbardSC/Project/HartreeFock/simulations/heatmap/Setup=A[128]/"
FilePath1 = RootFolder * "AF.txt"
FilePath2 = RootFolder * "FakeAF.txt"

Data1 = readdlm(FilePath1, ';')
Data2 = readdlm(FilePath2, ';')
"done"
```

```julia
Data1[1,:] .= [:t; :U; :V; :Lx; :β; :δ; :v; :Q; :ΔT]
Data2[1,:] .= [:t; :U; :V; :Lx; :β; :δ; :v; :Q; :ΔT]
"done"
```

```julia
DF1 = DataFrame(Data1[2:end,:], Data1[1,:])
DF2 = DataFrame(Data2[2:end,:], Data2[1,:])
"done"
```

```julia
V=0.0
δ=0.0
df1 = DF1[(DF1.V.==V) .& (DF1.δ.==δ), :]
df2 = DF2[(DF2.V.==V) .& (DF2.δ.==δ), :]
xx = df1.U
yy1 = [v["m"] for v in eval.(Meta.parse.(df1.v))]
yy2 = [v["m"] for v in eval.(Meta.parse.(df2.v))]
plot(
    xx,yy1,
    xlabel="U",
    ylabel="m",
    title="V=$V, δ=$δ, L=$(Int64(df1.Lx[1])), β=$(Int64(df1.β[1]))",
    label="t: ren"
)
plot!(
    xx,yy2,
    label="t: not ren"
)
```
